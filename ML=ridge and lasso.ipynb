```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('mtcars.csv')
df
```

```python
# head
```

```python
df.head()
```

```python
# drop categorical variable
```

```python
df.drop(['model'],axis=1,inplace=True)
df
```

```python
# create feature and target variable
```

```python
x=df.drop(columns='hp',axis=1)
y=df['hp']
```

```python
x.head()
```

```python
y.head()
```

```python
# splitting the data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
# fit the data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
# r2 score
```

```python
from sklearn.metrics import r2_score
```

```python
# r2 score for train data(relation)
```

```python
x_pred_train = reg.predict(x_train)
r2_score(y_train,x_pred_train)
```

```python
# r2 score for test data
```

```python
x_pred_test = reg.predict(x_test)
r2_score(y_test,x_pred_test)
```

# buildind a ridge model

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge()
ridge.fit(x_train,y_train)
```

```python
# predict
```

```python
x_pred_ridge_test = ridge.predict(x_test)
r2_score(y_test,x_pred_ridge_test)# relation how much value varry actual and test
```

# building a lasso model

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso()
lasso.fit(x_train,y_train)
```

```python
x_pred_lasso_test = lasso.predict(x_test)
r2_score(y_test,x_pred_lasso_test)
```
